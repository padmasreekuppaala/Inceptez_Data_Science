## Handling missing values in Demand Transaction csv file

```python
# Reading csv file
dTxn = pd.read_csv('Demand_txn_updated.csv')
```

```python
# Handling the missing values in column Percent_Quantity_With_Discount with strategy = mean
dTxn['Percent_Quantity_With_Discount'].fillna(dTxn['Percent_Quantity_With_Discount'].mean(),inplace=True)
```

```python
# Checking if all the null values are replaced by mean
dTxn['Percent_Quantity_With_Discount']
```

```python
# One hot encode the columns Mapped_Sales_Type getting dummies
dTxn['Mapped_Sales_Type'].value_counts()
```

```python
mapped = pd.get_dummies(dTxn['Mapped_Sales_Type'],prefix='MapSales',dummy_na=False)
dTxn = dTxn.drop('Mapped_Sales_Type',1)

# ignoring warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
dTxn.head()
```

```python
#  Label binarize the column Day
from sklearn.preprocessing import LabelBinarizer
lb_day = LabelBinarizer()
```

```python
lb_result = lb_day.fit_transform(dTxn['Day'])
day = pd.DataFrame(lb_result,columns=lb_day.classes_).tail()
```

```python
# concating the dataframe with the newly formed columns
pd.concat([dTxn,day],axis=1).tail()
```

```python
# Label encode the column City
dTxn.City.value_counts()
```

```python
# importing label encoder
from sklearn.preprocessing import LabelEncoder
lb_city = LabelEncoder()
dTxn['City_Code'] = lb_city.fit_transform(dTxn['City'])
```
