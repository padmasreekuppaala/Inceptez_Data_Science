```python
#Find the most expensive car company name
df.loc[df['price'] == df.price.max()]
```

```python
#Count total cars per company
df['make'].value_counts()
```

```python
#Find each companyâ€™s highest price car
df.groupby(['make'])['price'].agg('max')
```

```python
#Sort all cars by Price column
df.sort_values(by=['price'])
```

```python
#Group by make, fuel-type and body-style to calculate the average of city-mpg and highwaympg (miles per gallon)
df.groupby(['make', 'fuel_type'])['city-mpg'].avg()
df.groupby(['make', 'fuel_type'])['highway_mpg'].avg()
```
